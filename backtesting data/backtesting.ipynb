```python
import pandas as pd
import numpy as np
import ta
from datetime import datetime

# Convert the date series to date object
def Bollinger_band(data):
    data['date'] = pd.to_datetime(data['date'])

    # Calculate the ratio series
    #data['close'] = data['High'] / data['low']

    # Apply Bollinger Band indicator
    

    # Initialize variables for tracking trades
    entry_date = None
    entry_price = None
    exit_date = None
    exit_price = None
    trade_type=None
    trades = []

    # Perform trading strategy
    for i in range(len(data)-1):
        if data['signal'].iloc[i]=='Buy':
                entry_date = data['date'].iloc[i]
                entry_price = data['close'].iloc[i]
                exit_date = data['date'].iloc[i+1]
                exit_price = data['close'].iloc[i+1]
                trade_type='Buy'
        elif data['signal'].iloc[i]=='Short':
                entry_date = data['date'].iloc[i-1]
                entry_price = data['close'].iloc[i-1]
                exit_date = data['date'].iloc[i]
                exit_price = data['close'].iloc[i]
                trade_type='Short'
        trades.append([entry_date, entry_price, exit_price, exit_date,trade_type])
        entry_date = None
        entry_price = None
        exit_date = None
        exit_price = None
        trade_type = None

    # Create DataFrame for trades
    trades_df = pd.DataFrame(trades, columns=['Entry_date', 'Entry_price', 'Exit_price', 'Exit_date','Type'])
    return trades_df

# Read the data from CSV file
data5  = pd.read_csv('data_5min_Buy_short.csv')
data15 = pd.read_csv('data_15min_Buy_short.csv')
data30 = pd.read_csv('data_30min_Buy_short.csv')
data1h = pd.read_csv('data_1h_Buy_short.csv')
data4h = pd.read_csv('data_4h_Buy_short.csv')

# Apply Bollinger Bands strategy and get trades
result5  = Bollinger_band(data5)
result15 = Bollinger_band(data15)
result30 = Bollinger_band(data30)
result1h = Bollinger_band(data1h)
result4h = Bollinger_band(data4h)

# Write trades to CSV files
result5.to_csv('data_5min_Buy_short_bollinger.csv', index=False)
result15.to_csv('data_15min_Buy_short_bollinger.csv', index=False)
result30.to_csv('data_30min_Buy_short_bollinger.csv', index=False)
result1h.to_csv('data_1h_Buy_short_bollinger.csv', index=False)
result4h.to_csv('data_4h_Buy_short_bollinger.csv', index=False)
```
