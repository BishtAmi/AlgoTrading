```python
import pandas as pd
import pandas_ta as ta
import mplfinance as mpf

# Calculate Supertrend using pandas_ta
def supertrend_func(data):
    supertrend = ta.supertrend(high=data['High'], low=data['Low'], close=data['Close'], length=10, multiplier=3)

    # Generate buy/short signals based on Supertrend values
    signals = [''] * len(data)
    for i in range(1, len(data)):
        if data['Close'][i] > supertrend['SUPERT_10_3.0'][i]:
            signals[i] = 'Buy'
        else:
            signals[i] = 'Short'

    # Create a new DataFrame with the desired columns
    result = pd.DataFrame({
        'date': data['Date_time'],
        'open': data['Open'],
        'High': data['High'],
        'low': data['Low'],
        'close': data['Close'],
        'signal': signals
    })
    return result
data5=pd.read_csv('data_resampled_5min.csv')
data15 =pd.read_csv('data_resampled_15min.csv')
data30 =pd.read_csv('data_resampled_30min.csv')
data1h =pd.read_csv('data_resampled_1hour.csv')
data4h =pd.read_csv('data_resampled_4hour.csv')

result5 = supertrend_func(data5)
result15 = supertrend_func(data15)
result30 = supertrend_func(data15)
result1h = supertrend_func(data1h)
result4h = supertrend_func(data4h)

result5.to_csv('data_5min_Buy_short.csv')
result15.to_csv('data_15min_Buy_short.csv')
result30.to_csv('data_30min_Buy_short.csv')
result1h.to_csv('data_1h_Buy_short.csv')
result4h.to_csv('data_4h_Buy_short.csv')
```
